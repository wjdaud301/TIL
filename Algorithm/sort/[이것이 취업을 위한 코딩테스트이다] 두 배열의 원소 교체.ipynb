### * 핵심 아이디어 : 매번 배열 A에서 가장 작은 원소를 골라서, 배열 B에서 가장 큰 원소와 교체한다!!

```python
n,k = map(int,input().split())
a = list(map(int,input().split()))  
b = list(map(int,input().split()))
a.sort()
b.sort(reverse=True)

cnt = 0
for i in range(n):
    if k == 0: break
    if a[cnt] < b[cnt]:
        a[cnt] = b[cnt]
        k -= 1
    else:
        continue
    cnt+=1

print(sum(a))
```
