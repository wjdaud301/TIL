```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.128 Safari/537.36"}

 # 7개 정보를 빈 리스트 만들기 
number = []
name = []
position = []
age = []
nation = []
team = []
value = []

for i in range(1,3):
    url = f"https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop?ajax=yw1&page={i}"
    res = requests.get(url,headers = headers)
    res.status_code

    soup = BeautifulSoup(res.content, 'lxml')
    
    # 선수들의 정보가 담긴 태그와 클래스 찾기
    player_info = soup.find_all('tr', class_=['odd','even'])

    # player_info에서 'td'태그만 모두 찾기
    for info in player_info:
        player = info.find_all("td")

        # 해당 정보를 찾아서 각 리스트에 append로 추가하기
        number.append(player[0].text)
        name.append(player[3].text)
        position.append(player[4].text)
        age.append(player[5].text)
        nation.append(player[6].img['alt'])
        team.append(player[7].img['alt'])
        value.append(player[8].b.text)
        
    time.sleep(1)


# pd.DataFrame()으로 저장하기 
df = pd.DataFrame({
    "number":number,
    "name":name,
    "position":position,
    "age":age,
    "nation":nation,
    "team":team,
    "value":value
})
df
```

```python
#df.to_csv()로 csv 파일로 저장하기
df.to_csv('transfermark25.csv',index = False)
```

## 다른 방식으로 DataFrame 만들기(리스트 안에 리스트)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.128 Safari/537.36"}

player_list=[]

for i in range(1,3):
    url = f"https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop?ajax=yw1&page={i}"
    res = requests.get(url,headers = headers)
    res.status_code

    soup = BeautifulSoup(res.content, 'lxml')
    
    # 선수들의 정보가 담긴 태그와 클래스 찾기
    player_info = soup.find_all('tr', class_=['odd','even'])

    # player_info에서 'td'태그만 모두 찾기
    for info in player_info:
        player = info.find_all("td")

        # 해당 정보를 찾아서 각 리스트에 append로 추가하기
        number = player[0].text
        name = player[3].text
        position =player[4].text
        age = player[5].text
        nation = player[6].img['alt']
        team = player[7].img['alt']
        value = player[8].b.text
        
        player_list.append([number,name,position,age,nation,team,value])
    
    time.sleep(1)


# pd.DataFrame()으로 저장하기 
df = pd.DataFrame(player_list, columns = ['등수','이름','포지션','나이','나라','소속','연봉가치'])
df
```

```python
#df.to_csv()로 csv 파일로 저장하기
df.to_csv('transfermark50.csv',index = False)
```

## 데이터 전처리(pre-processing)

```python
# 연봉가치에서 불필요한 문자는 없애고 데이터 타입을 숫자형으로 바꾸기
# €160.00m -> 160.00
# 열에 대한 전처리를 모두 적용하기 위해서 apply함수와 lambda함수를 통해 한꺼번에 전처리
import re
df['연봉가치'] = df['연봉가치'].apply(lambda x : float(x[1:len(x)-1]))
df.head()
```

## 칼럼 생성과 삭제

```python
# 컬럼 생성 -> 시장가치단위 유로를 한화로
df['시장가치(억)'] = df['연봉가치']*13
df.head()
```

```python
# 칼럼 삭제 df.drop(columns = ['칼럼이름'])
df.drop(columns =['연봉가치'], inplace=True)
```
