```python
# slicing에 대한 결과에서 원하는 컬럼을 빼고 싶으면 drop함수를 이용
df3 = df1.iloc[:,6:-2].drop(columns=['Unnamed: 7','판매'])
df3
```

```python
df4 = pd.DataFrame(df3.set_index(['제품명']).stack()).reset_index()
df4
```

```python
# 컬럼의 이름을 바꾸기 위해서 rename함수를 이용 (딕셔너리 형태)
df5 = df4.rename(columns ={'level_1':'날짜',0:'재고량'}) 
```

```python
df6 = df1[['카테고리명','자재그룹','자재그룹명','제품코드','제품명','안전재고',' 분류']]
```

```python
df6.head(2)
```

* Table Join
    - right / left
    - inner
    - outer

```python
df5.head(2)
```

```python
df7 = pd.merge(df6,df5,how='left',on = '제품명')
df7.to_excel('./data/total_data.xlsx')
```

```python
df7['날짜(전처리)'] = pd.to_datetime(df7['날짜'])
```

```python
df7['공급월'] = df7['날짜(전처리)'].dt.month
```

```python
# 피벗테이블도 엑셀파일로 저장이 가능하다

pd.pivot_table(data=df7,index='공급월'
               ,values='재고량',aggfunc='sum').to_excel('result.xlsx')
```

```python
# 공급월에 따른 재고량 시각화

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rc('font', family='Malgun Gothic')
```

```python
sns.barplot(data=df7, x='공급월', y='재고량', ci = None, estimator =sum)
plt.savefig('img_result.png')
```
