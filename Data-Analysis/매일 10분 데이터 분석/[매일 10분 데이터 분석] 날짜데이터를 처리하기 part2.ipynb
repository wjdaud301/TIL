```python
import pandas as pd
```

```python
df2 = pd.read_csv('./data./Data02.csv')
print(df2.shape)
df2.head()
```

```python
# apply : pandas에서 사용하는 함수적용식

df2['배송시작일(datetime)'] = pd.to_datetime(df2['배송시작일'].apply(lambda x: x[0:10]))
df2['배송완료일(datetime)'] = pd.to_datetime(df2['배송완료일'].apply(lambda x: x[0:10]))
```

```python
df2.head()
```

```python
# 시간의 간격 나타내기 (시간과 시간을 빼면 그만큼 차이를 발생)

df2['배송시간'] = df2["배송완료일(datetime)"] -df2["배송시작일(datetime)"]
df2.head()
```

```python
df2['배송시간'].describe()
```

```python
# 0000-00-00이 아닌 형태를 처리

pd.to_datetime(df2['주문일'].apply(lambda x: str(x)[0:4] + '-' + str(x)[4:6] + '-' + str(x)[6:]))
```
