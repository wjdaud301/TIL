```python
import sqlite3
import pandas as pd
```

```python
# database를 접근하기 위해서 connection함수를 이용 (들어갈 수 있는 key)

connection = sqlite3.connect('./data/sqlite/sqlite-tools-win32-x86-3350500/chinook.db')
```

* Select : 데이터를 찾을 때! 데이터 테이블을 불러올 때
* Insert : 데이터를 집어넣을 때!
* Update : 데이터를 갱신할 때!
* Delete : 데이터를 삭제할 때!

```python
query = "SELECT * FROM sqlite_master;" # sql문법
```

```python
# sql ( query, connection ) 순서 

database_table = pd.read_sql_query(query,connection)
database_table['sql']
```

```python
# 접근 후 pandas로 데이터 조작

query = "SELECT * FROM 'customers';"
df_customers = pd.read_sql_query(query,connection)
df_customers.head()
```

```python
# database file -> excel file로 저장

df_customers.to_excel('./data/customer_file.xlsx')
```

```python
query = "SELECT * FROM 'invoices';"
df_invoices = pd.read_sql_query(query,connection)
df_invoices.head()
```

```python
# pandas로 조건데이터 불러오기

cond1 = (df_invoices['BillingCountry'] == 'Germany')
df_invoices.loc[cond1].head()
```

```python
# sql문법으로 조건데이터 불러오기

query = "SELECT * FROM 'invoices' WHERE BillingCountry = 'Germany';"
df_invoices = pd.read_sql_query(query,connection)
df_invoices.head()
```
