## 상위 10개 데이터를 뽑아서 시각화하기

### 1. bank 속성중에서 빈도수가 많은 값으로 정렬
### 2. 정렬된 값들 중에 Index(은행명)를 리스트로 저장
### 3. 리스트중에 10개의 데이터를 뽑아 countplot order 인자로 넘겨준다.

```python
order_list = df1['bank'].value_counts().index.tolist()
```

```python
plt.figure(figsize=[10,5])
sns.countplot(data=df1, x= 'bank', order= order_list[0:10])
```
